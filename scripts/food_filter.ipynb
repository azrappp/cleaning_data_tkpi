```python
import pandas as pd
import re

# ============================================================
# 1. Import final dataframe
# ============================================================

input_file = "final_data_frame.csv"

df = pd.read_csv(input_file)

print("Total rows before filtering:", len(df))
print("Columns:")
print(df.columns.tolist())


# ============================================================
# 2. Basic text preparation
# ============================================================

df["food_name_clean"] = (
    df["food_name"]
    .astype(str)
    .str.strip()
    .str.lower()
)

df["category_code"] = (
    df["category_code"]
    .astype(str)
    .str.strip()
    .str.upper()
)


# ============================================================
# 3. Define unwanted foods
# ============================================================

exclude_keywords = [
    # Not practical for direct menu recommendation
    "tepung",
    "maizena",
    "hunkwe",

    # Sugar / sweeteners, not suitable for DM
    "gula",
    "sirup",
    "madu",
    "selai",
    "jam",

    # Processed / high sodium foods
    "ikan asin",
    "asin kering",
    "kornet",
    "sosis",
    "ham",
    "sardencis",
    "telur bebek asin",

    # Not recommended / less suitable
    "daging babi",
    "otak",
    "usus",
    "babat",
    "ginjal",

    # Uncommon vegetables / less menu-practical
    "daun pangi",
    "putri malu",
    "andaliman",
    "akar tonjong"
]

exclude_pattern = "|".join([re.escape(x) for x in exclude_keywords])

df["is_excluded_keyword"] = df["food_name_clean"].str.contains(
    exclude_pattern,
    na=False
)


# ============================================================
# 4. Filter only ready foods
# ============================================================

ready_food_df = df[
    df["is_excluded_keyword"] == False
].copy()

ready_food_df["is_menu_ready"] = True


# ============================================================
# 5. Save new CSV
# ============================================================

ready_output_file = "ready_food_data_frame.csv"

ready_food_df.to_csv(ready_output_file, index=False)

print("Total rows after filtering:", len(ready_food_df))
print("Excluded rows:", len(df) - len(ready_food_df))
print("Saved file:", ready_output_file)


# ============================================================
# 6. Preview result
# ============================================================

display(
    ready_food_df[
        [
            "food_name",
            "category_code",
            "urt",
            "gram_per_portion",
            "is_menu_ready"
        ]
    ].head(30)
)
```
